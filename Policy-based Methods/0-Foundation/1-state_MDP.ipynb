### The construction of MDP
(1) Statement Space: S = {s}<br>
(2) Action Space: A = {left, right}<br>
(3) pi(s) = 0.5left+0.5right<br>
(4) pi'(s)=always left<br>
(5) reward: left=10, right=3<br>
(6) no future rewards

```python
import numpy as np

# MDP setup
rewards = {'left':10, 'right':3}
gamma = 0.0 # no future rewards

# Policy π (uniform)
pi = {"left": 0.5, "right": 0.5}

# Policy π'
pi_prime = {"left": 1.0, "right": 0.0}
```

```python
# Compute Jπ
J_pi = sum(pi[a] * rewards[a] for a in pi) # the expected reward using policy pi
print("J_pi =", J_pi)

# Compute Qπ
Q_pi = {a: rewards[a] for a in rewards}
print("Q_pi =", Q_pi)
```

```python
# Compute Aπ
A_pi = {a: Q_pi[a] - J_pi for a in rewards}
print("A_pi =", A_pi)

# Compute Jπ'
J_pi_prime = rewards["left"]  # π' always picks left
print("J_pi_prime =", J_pi_prime)
```

```python
# Difference Lemma RHS: sum γ^k Eπ'[Aπ]
DL_RHS = A_pi["left"]  # only left is chosen
print("Difference Lemma RHS =", DL_RHS)

# Real difference
print("Actual difference =", J_pi_prime - J_pi)
```
